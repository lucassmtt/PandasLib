```python
from datetime import datetime, timedelta
base = datetime.today()
base_times = 900
data_list = [base - timedelta(days=x) for x in range(base_times)]
```

```python
data_list
```

```python
freq = Series(
    [randint(1, 10_000) for _ in range(base_times)],
    index=data_list
)
```

```python
freq_deste_mes = freq.iloc[0: 30]
```

```python
freq_deste_mes
```

```python
g = Grouper(freq='1Y')
```

```python
freq.groupby(g).count()
```

```python
freq.groupby(g).count().plot(kind='barh')
```

```python
print(freq.groupby(g).sum())
freq.groupby(g).sum().plot(kind='barh')
```
